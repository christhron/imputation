```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Reading the data

```python
df = pd.read_csv(os.path.abspath('../data/Nal_filtre_FLX_NEE_01012011_31122017.csv'), encoding = "ISO-8859-1", sep = ';')
df.columns = df.iloc[0] # fix the col names to be the 1st col
df = df.iloc[1:] # fix the col names to be the 1st col
```

```python
df.head(3)
```

---

### Preprocessing

```python
X = df.iloc[:,2:].values
X = X.astype(float)
print(X.shape)
```

#### Tensor

```python
X = X[:365 * (len(X)//365) ,] # we are removing the extra days
print(X.shape)
X_tensor = np.reshape(X, (7, -1,8 )) 
print(X_tensor.shape) # (t, rows, cols)
print(X_tensor[0][0])
```

---

### 1. Seasonality

```python
Xmean = np.nanmean(X_tensor, axis = 0) # so we took the avrage across 7 years
print(Xmean.shape)
print(Xmean[0])
```

```python
plt.plot(Xmean[::48,0]) # plotting NEE
```

---

### 2. Smoothing

```python
from astropy.convolution import convolve, Box1DKernel
```

```python
Xsmoothed = np.zeros(Xmean.shape)
for i in range (8):
    box = Box1DKernel(5)
    smoothed_gauss = convolve(Xmean[:,i], box)
    Xsmoothed[:,i] = smoothed_gauss
    print(smoothed_gauss.shape)
    plt.plot(Xmean[0:1000,i])
    plt.plot(smoothed_gauss[0:1000], 'r:')
    plt.show()
```

```python
print(Xsmoothed)
```

---

### 3. Removing Seasonality

```python
XClean_tensor = X_tensor - Xsmoothed
print(XClean_tensor.shape)
print(XClean_tensor[0][0])
```

---

### 4. Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xstand = scaler.fit_transform( np.reshape(XClean_tensor,(-1,8)) )
```

```python
Xstand.shape
```

---

### 5. Outliers Removal

```python
Xout = np.abs(Xstand)>3 # we tried with 4, but the number was still so big
Xstand[Xout] = float("nan")
Xstand[Xout] = float("nan")
```
